```python
import pandas as pd

df = pd.read_excel("data/wine_sales.xlsx")

new_columns = ['id', 'date_int', 'product_name', 'price', 'sales', 'reviews', 'brand', 'searches']
df.columns = new_columns

df.head(4)
```

```python
df.info()
```

```python
df.describe()
```

```python
filas_con_almenos_un_na = df[df.isna().any(axis=1)]
filas_todas_na = df[df.isna().all(axis=1)]
print(f"Filas con almenos un na: {len(filas_con_almenos_un_na)}")
print(f"Filas con todos los valores en na: {len(filas_todas_na)}")
print(f"total filas: {len(df)}")

```
